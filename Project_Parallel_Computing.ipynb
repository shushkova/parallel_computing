# Schrodinger Equation

Основным уравнением нерелятивистской квантовой механики является временное уравнение Шредингера:

$$ i \hbar \frac{\partial  \psi (x,y,z,t)}{\partial t} = \hat H \psi (x,y,z,t)$$

$$\hat H = \hat T + \hat U $$

$$\hat U = U $$

$$\hat T = \frac{\hat p ^2}{2m} ,\space  \hat p = -i \hbar \nabla $$

$$\hat H  =  \frac{\hat p ^2}{2m} + \hat U =  \frac{(-i \hbar \nabla)^2}{2m} + \hat U  = -\frac{\hbar^2}{2m} \Delta + \hat U $$


Подставляя, получим

$$i \hbar \frac{\partial \psi(x,y,z,t)}{\partial t} = - \frac{\hbar^2}{2m} \Delta \psi(x,y,z,t) + U(x,y,z,t) \psi(x,y,z,t) $$


## 1D Time-independent Schrodinger Equation (Стационарное уравнение Шредингера)

В квантовой механике существует класс задач о движении в силовых полях, для которых силовая функция $U(x,y,z,t)$ не зависит явно от времени, т.е.

$$U(x,y,z,t) = U(x,y,z)$$

Поскольку оператор гамильтона $\hat H $ в уравнении не зависит явно от времени, то волновую функцию следует искать в виде произведения двух функций:

$$\psi(x,y,z,t) = \psi(x,y,z) \cdot \psi (t)$$


Подставим теперь волновую функцию в уравнение:

$$ i \hbar \frac{\partial  \psi (x,y,z,t)}{\partial t} = \hat H \psi (x,y,z,t)$$

$$ i \hbar \frac {\partial }{\partial t}  (\psi(x,y,z) \cdot \psi (t)) = \hat H  (\psi(x,y,z) \cdot \psi (t) )$$

Let's say $x,y,z \rightarrow r $

$$ i \hbar \frac {\partial }{\partial t}  (\psi(r) \cdot \psi (t)) = \hat H  (\psi(r) \cdot \psi (t) )$$

$$ i \hbar \frac{1}{\psi(t)} \frac {\partial }{\partial t} \psi (t) =\frac{1}{\psi(r)} \hat H  \psi(r) $$

В уравнении выше левая часть зависит только от времени, а правая - только от координат. Выполнение этого равенства возможно лишь в том случае, если левая и правая части уравнения равны постоянной величине. Обозначим эту величину как $E$.

$$ i \hbar \frac{1}{\psi(t)} \frac {\partial }{\partial t} \psi (t) =\frac{1}{\psi(r)} \hat H  \psi(r) = E$$

Тогда

$$ i \hbar \frac{1}{\psi(t)} \frac {\partial }{\partial t} \psi (t)  = E $$

$$\frac{\partial \psi(t)}{\psi(t)} = \frac{E}{i \hbar} \partial t$$

$$ln\psi(t) = \frac{E}{i \hbar}t + lnC $$

$$\psi(t) = e^{\frac{E}{i \hbar}t} C$$

Так как $\psi(x,y,z,t) = \psi(x,y,z) \cdot \psi (t)$, и $\psi(x,y,z)$ тоже определяется с точностью до произвольной константы, нет смысла плодить произвольные константы, поэтому положим, что тут $C = 1$.

Тогда

$$\psi(t) = e^{\frac{E}{i \hbar}t}$$

Тогда запишем волновую функцию, как

$$\psi(r,t) = \psi(r) \cdot e^{\frac{E}{i \hbar}t}$$

Подставим это в уравнение Шредингера

$$i \hbar \frac{\partial \psi(r,t)}{\partial t} = - \frac{\hbar^2}{2m} \Delta \psi(r,t) + U(r,t) \psi(r,t) $$

$$i \hbar \frac{\partial }{\partial t}(\psi(r) \cdot e^{\frac{E}{i \hbar}t}) = - \frac{\hbar^2}{2m} \Delta (\psi(r) \cdot e^{\frac{E}{i \hbar}t})+ U(r) \psi(r) \cdot e^{\frac{E}{i \hbar}t} $$

$$i \hbar \cdot \psi(r) \frac{\partial e^{\frac{E}{i \hbar}t} }{\partial t} = - \frac{\hbar^2}{2m} \cdot e^{\frac{E}{i \hbar}t} \cdot \Delta \psi(r) + e^{\frac{E}{i \hbar}t} \cdot U(r) \psi(r)$$

$$i \hbar \cdot \psi(r) \cdot e^{\frac{E}{i \hbar}t} \cdot \frac{E}{i \hbar} = - \frac{\hbar^2}{2m} \cdot e^{\frac{E}{i \hbar}t} \cdot \Delta \psi(r) + e^{\frac{E}{i \hbar}t} \cdot U(r) \psi(r)$$

$$\psi(r) \cdot E = - \frac{\hbar^2}{2m}  \cdot \Delta \psi(r) +  U(r) \psi(r)$$

$$- \frac{\hbar^2}{2m}  \cdot \Delta \psi(r) +  U(r) \psi(r) -\psi(r) \cdot E = 0$$

$$\Delta \psi(r) +\psi(r) \cdot \frac{2m}{\hbar^2}(E - U(r) )= 0 -\text{  Стационарное Уравнение Шредингера}$$

### 1D case of Schrodinger Equation

$$\Delta \psi(r) +\psi(r) \cdot \frac{2m}{\hbar^2}(E - U(r) )= 0 -\text{  Стационарное Уравнение Шредингера}$$

Уравнение Шредингера в одномерном поле будет

$$\frac{\partial^2 }{\partial x^2} \psi(x) +\frac{2m}{\hbar^2}(E - U(x) ) \cdot  \psi(x)= 0$$

#### Прямоугольная потенциальная яма с бесконечными стенками

изображение.png

$$
U(x) = 
\begin{cases}
  0 &\text{при } 0 < x < L\\
  \infty &\text{при } x \leq 0 , x \geq L
\end{cases} 
 $$

Мы будем рассматривать случай $0 < x < L$, где $U(x) = 0$, т.к. вне этой области $\psi (x) = 0 $, т.е.

$$\text{ Граничные условия: } \psi(x=0) = \psi(x = L) = 0 $$

Таким образом решаем уравнение в свободном пространстве, где отсутствуют потенциалы, т.е. находим решение для свободной частицы.

$$\frac{\partial^2 }{\partial x^2} \psi(x) +\psi(x) \cdot \frac{2m}{\hbar^2}(E - U(x) )= 0$$

$$U(x) = 0 $$

Тогда уравнение будет иметь вид

$$\frac{\partial^2 \psi(x) }{\partial x^2} +\psi(x) \cdot \frac{2m}{\hbar^2}E= 0$$

##### Аналитическое решение

###### Вывод аналитического решения

Итак, наше уравнение выглядит так

$$\frac{\partial^2 \psi(x) }{\partial x^2} +\psi(x) \cdot \frac{2m}{\hbar^2}E= 0$$

Для простоты заменим $k^2 = \frac{2m}{\hbar^2}E$, тогда

$$\frac{\partial^2 \psi(x) }{\partial x^2} +\psi(x) \cdot k^2 = 0$$

$$\psi^{\prime\prime} + k^2 \psi = 0 $$

For convenience, I will rewrite it as:

$$y^{\prime\prime} + k^2 y = 0 $$

Составляем характеристическое уравнение:

$$\lambda^2 + k^2 = 0$$

$$ \lambda^2 =- k^2 $$

$$\lambda = \pm \sqrt{-k^2} =  \pm \sqrt{-1} \cdot k = \pm ik$$

$$y = C_1 cos{kx} + C_2 sin{kx} $$

Итак, возвращаемся к $\psi$

$$\psi^{\prime\prime} + k^2 \psi = 0 $$

$$\psi(x) = C_1 cos{kx} + C_2 sin{kx}, \text{ where } k = \frac{\sqrt{2mE}}{\hbar}$$

Используя граничные условия, найдем производные постоянные $C_1$ и $C_2$

$$\psi(x = 0) = C_1 cos({k \cdot 0}) + C_2 sin({k \cdot 0}) = C_1 $$
$$\psi(x = 0) = 0 \rightarrow C_1 = 0$$

Тогда 

$$\psi(x) = C_2 sin{kx} $$

Используем второе граничное условие

$$\psi(x=L) = C_2 sin({kx}) = C_2 sin({k \cdot L}) = 0$$

$C_2 $ не может быть нулем, поэтому

$$ sin({k \cdot L}) = 0 $$

$$ kL = \pi n,  n \in Z$$

$$ k = \frac{\pi n}{L} , n = 1,2,3...., \text{ т.е.} n \in Z$$


Это нам говорит о том, что энергия принимает дискретные значения, т.е.собственные функции существуют лишь при определѐнных значениях полной энергии Е,
называемых **собственными значениями** энергии. Совокупность собственных значений Е образуют **энергетический спектр** частицы.

$$E_n = \frac{\hbar^2 k^2}{2m} = \frac{\hbar^2 \pi^2 n^2}{2m}, n=1,2,...$$

Чтобы найти $C_2$, воспользуемся условием нормировки волновой функции

$$ \int_0^L |\psi_n(x)|^2 dx = 1$$ 

$$\int_0^L |\psi_n(x)|^2 dx  = \int_0^L |C_2 sin({kx})|^2 dx =   \int_0^L C_2^2 sin^2({kx}) dx = C_2^2 \int_0^L sin^2({kx}) dx$$

Давайте теперь решим этот интеграл

$$\int_0^L sin^2({kx}) dx= \frac{1}{k} \int_0^L sin^2({kx}) d(kx)  = \frac{1}{k} \int_0^L \frac{1-cos(2kx)}{2} d(kx) = $$

$$=  \frac{1}{k} \int_0^L (\frac{1}{2} - \frac{1}{2} cos(2kx)) d(kx) = \frac{1}{k} \left(\frac{1}{2}kx|^L_0 -\frac{1}{2} \int_0^L  cos(2kx)) d(kx) \right)  = $$

$$ = \frac{1}{k} \left(\frac{1}{2}kx|^L_0 -\frac{1}{2} \cdot \frac{1}{2} \int_0^L  cos(2kx)) d(2kx) \right) = \frac{1}{k} \left(\frac{1}{2}kx|^L_0 -\frac{1}{4} sin(2kx)|_0^L \right)  = \frac{1}{2} (L-0) - \frac{1}{4k} (sin(2kL) - sin0) = \frac{1}{2}L - \frac{1}{4k} sin(2kL)$$

$$ k = \frac{\pi n}{L} , n = 1,2,3...., \text{ т.е.} n \in Z$$

$$ \frac{1}{2}L - \frac{1}{4k} sin(2kL) = \frac{1}{2}L - \frac{L}{4 \pi n} sin(2 \cdot \frac{\pi n}{L} \cdot L) = \frac{1}{2}L - \frac{L}{4 \pi n} sin(2 \pi n) = \frac{1}{2}L$$

Возвращаемся к исходному интегралу


$$ C_2^2 \int_0^L sin^2({kx}) dx = C_2^2 \cdot \frac{1}{2}L$$

И так как по условию нормировки

$$ \int_0^L |\psi_n(x)|^2 dx = 1$$ 

То

$$C_2^2 \cdot \frac{1}{2}L = 1$$

$$C_2 = \sqrt{\frac{2}{L}} $$

Тогда волновая функция будет

$$\psi_n (x) =  \sqrt{\frac{2}{L}} sin\left( \frac{\pi n}{L} x\right), \text{ при } E_n =  \frac{\hbar^2 \pi^2 n^2}{2m},  n = 1,2,3...., \text{ т.е.} n \in Z$$

###### Построение графиков

Итак, аналитическим решением является

$$\psi_n (x) =  \sqrt{\frac{2}{L}} sin\left( \frac{\pi n}{L} x\right), \text{ при } E_n =  \frac{\hbar^2 \pi^2 n^2}{2m},  n = 1,2,3...., \text{ т.е.} n \in Z$$

Для простоты поставим значение $\hbar = m = L = 1$, так что

$$\psi_n (x) =  \sqrt{2} sin\left(\pi n x\right), \text{ при } E_n =  \frac{ \pi^2 n^2}{2},  n = 1,2,3....$$

##### Численное решение

Итак,

Уравнение Шредингера в одномерном поле будет

$$\frac{\partial^2 }{\partial x^2} \psi(x) +\frac{2m}{\hbar^2}(E - U(x) ) \cdot  \psi(x)= 0$$

$$\text{ Граничные условия: } \psi(x=0) = \psi(x = L) = 0 $$

Перепишем уравнение Шредингера в другой вид

$$ - \frac{\hbar^2}{2m} \frac{\partial^2  \psi(x) }{\partial x^2}+ U(x) \psi(x) = E\psi(x)  $$

$$ - \frac{1}{2} \frac{\partial^2  \psi(x) }{\partial x^2}+ \frac{m}{\hbar^2} U(x) \psi(x) = \frac{m}{\hbar^2} E\psi(x) $$

Здесь у нас $0 < x < L$, переведем все в диапазон от $0$ до $1$ путем замены $y = \frac{x}{L}$

$$y = \frac{x}{L}$$

$$dy^2 = d\left( \left(\frac{x}{L} \right)^2 \right) = d\left(\frac{x^2}{L^2} \right) $$

$$t = x^2 \rightarrow d\left(\frac{t}{L^2} \right) = \frac{1}{L^2}dt$$

$$dt = L^2 d\left(\frac{t}{L^2} \right)$$

$$dx^2 =   L^2 d\left(\frac{x^2}{L^2} \right) =   L^2 d\left( \left(\frac{x}{L} \right)^2 \right) = L^2 dy^2$$

Итак, делаем замену

$$ - \frac{1}{2} \frac{\partial^2  \psi(y) }{L^2 \partial y^2}+ \frac{m}{\hbar^2} U(y) \psi(y) = \frac{m}{\hbar^2} E\psi(y)$$

$$ - \frac{1}{2} \frac{\partial^2  \psi(y) }{\partial y^2}+ L^2\frac{m}{\hbar^2} U(y) \psi(y) = L^2 \frac{m}{\hbar^2} E\psi(y)$$

Воспользуемся методом конечных разностей, в которой производную второго порядка можно представить в виде

$$\frac{d^2f}{dx^2} = \frac{f_{j+1}-2f_j+f_{j-1}}{\Delta x^2}, \text{где } f_j \equiv f(j\Delta x)$$

Перепишем уравнение Шредингера

$$ - \frac{1}{2} \frac{\psi_{j+1}-2\psi_j+\psi_{j-1}}{\Delta y^2} + L^2\frac{m}{\hbar^2} U_j \psi_j = L^2 \frac{m}{\hbar^2} E\psi_j$$

$$- \frac{1}{2\Delta y^2}\psi_{j+1} + \left( \frac{1}{\Delta y^2} + L^2\frac{m}{\hbar^2} U_j\right) \psi_j - \frac{1}{2\Delta y^2}\psi_{j-1} = L^2 \frac{m}{\hbar^2} E\psi_j$$

Можно понять, что это СЛАУ

$$
\begin{cases}
  - \frac{1}{2\Delta y^2}\psi_{2} + \left( \frac{1}{\Delta y^2} + L^2\frac{m}{\hbar^2} U_j\right) \psi_1 - \frac{1}{2\Delta y^2}\psi_{0} = L^2 \frac{m}{\hbar^2} E\psi_1 \\
   - \frac{1}{2\Delta y^2}\psi_{3} + \left( \frac{1}{\Delta y^2} + L^2\frac{m}{\hbar^2} U_j\right) \psi_2 - \frac{1}{2\Delta y^2}\psi_{1} = L^2 \frac{m}{\hbar^2} E\psi_2 \\
  \ldots \\
   - \frac{1}{2\Delta y^2}\psi_{N} + \left( \frac{1}{\Delta y^2} + L^2\frac{m}{\hbar^2} U_j\right) \psi_{N-1} - \frac{1}{2\Delta y^2}\psi_{N-2} = L^2 \frac{m}{\hbar^2} E\psi_{N-1} \\
\end{cases} 
 $$

Перепишем в виде матрицы
$$
\begin{bmatrix} 
\left( \frac{1}{\Delta y^2} + L^2\frac{m}{\hbar^2} U_1\right)  & - \frac{1}{2\Delta y^2} & 0 & \ldots & 0 & 0\\
- \frac{1}{2\Delta y^2} & \left( \frac{1}{\Delta y^2} + L^2\frac{m}{\hbar^2} U_2\right)  & - \frac{1}{2\Delta y^2} & 0 & \ldots & 0 \\
0 & - \frac{1}{2\Delta y^2} & \left( \frac{1}{\Delta y^2} + L^2\frac{m}{\hbar^2} U_3\right)  & - \frac{1}{2\Delta y^2} & \ldots & 0  \\
 & & \ldots & \ldots &  &   \\
 & & \ldots & \ldots &  &   \\
0  && \ldots  & 0 &  - \frac{1}{2\Delta y^2} & \left( \frac{1}{\Delta y^2} + L^2\frac{m}{\hbar^2} U_{N-1}\right) 
\end{bmatrix} 
\begin{bmatrix} \psi_1 \\ \psi_2 \\ \vdots \\ \vdots \\ \\ \\ \psi_{N-1} \end{bmatrix}
= L^2 \frac{m}{\hbar^2} E \begin{bmatrix} \psi_1 \\ \psi_2 \\ \vdots \\ \vdots \\ \\ \\ \psi_{N-1} \end{bmatrix}
$$


Now let's define $N$, $\Delta y$ and y

$$y \in [0,1] $$

$$N \cdot \Delta y = 1 \rightarrow \Delta y = \frac{1}{N}$$

```python
N = 2000
dy = 1/N
y = np.linspace(0,1,N+1) # y from 0 to N
```

For simplicity and as from analytical part, we will put $\hbar = m = L = 1$

Тогда главная диагональ будет выглядеть так:  $\left( \frac{1}{\Delta y^2} + U_j\right) $ 

Now we need to define this expression $\left( \frac{1}{\Delta y^2} + U_j\right) $  on the main diagonal of our matrix

```python
def U(y):
  return y * 0.

def main_diag(dy,y):
  return 1/dy**2 + U(y)[1:-1]
```

Построим нашу матрицу А

$$ A =
\begin{bmatrix} 
\left( \frac{1}{\Delta y^2} + L^2\frac{m}{\hbar^2} U_1\right)  & - \frac{1}{2\Delta y^2} & 0 & \ldots & 0 & 0\\
- \frac{1}{2\Delta y^2} & \left( \frac{1}{\Delta y^2} + L^2\frac{m}{\hbar^2} U_2\right)  & - \frac{1}{2\Delta y^2} & 0 & \ldots & 0 \\
0 & - \frac{1}{2\Delta y^2} & \left( \frac{1}{\Delta y^2} + L^2\frac{m}{\hbar^2} U_3\right)  & - \frac{1}{2\Delta y^2} & \ldots & 0  \\
 & & \ldots & \ldots &  &   \\
 & & \ldots & \ldots &  &   \\
0  & \ldots  & 0 &  - \frac{1}{2\Delta y^2} & \left( \frac{1}{\Delta y^2} + L^2\frac{m}{\hbar^2} U_{N-1}\right)  & - \frac{1}{2\Delta y^2} 
\end{bmatrix} 
$$

$$A \psi = b \psi $$

###### Нахождение eigenvalues через QR algorithm 

```python
def qr_algorithm(A_init, num_iter): # 3 pts
    # enter your code here

    convergence = np.array([])
    for k in range(num_iter):
      q, rmat = np.linalg.qr(A_init)
      Ak = np.dot(rmat,q)
      A_init = Ak
 
      convergence = np.append(convergence,np.linalg.norm(np.tril(Ak, -1),ord=np.inf))
    return Ak, convergence
```

```python
d = main_diag(dy,y)
e = -1/(2*dy**2) * np.ones(len(d)-1)
diagonals = [d, e, e]
A_init = diags(diagonals, [0, -1, 1]).toarray()
```

```python
num_iter = 400
Ak,convergence = qr_algorithm(A_init, num_iter)
```

```python
Ak.diagonal(0)
```

```python
import matplotlib.pyplot as plt
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

ax1.plot(convergence)
ax1.set_xlabel('Number of iteration', fontsize=20)
ax1.set_ylabel('Convergence', fontsize=20)
ax1.tick_params(labelcolor='b', labelsize=15, width=3)
# plt.yscale('log')
ax1.set_title('Convergence of QR algorithm', fontsize=20)
ax1.grid()

ax2.spy(Ak, precision=1e-7)
ax2.set_title('Output matrix', fontsize=20)

plt.show()
```

## 1D Time-dependent Schrodinger Equation (Нестационарное уравнение Шредингера)

### Time dependent Schrodinger equation

Time dependent Schrodinger equation for the complex wave function $\psi(x,t)$, with $h^{2} / 2m = 1$ ...



$$ i \frac{\partial \psi(x,t)}{\partial t} = - \frac{\partial^{2} \psi(x,t)}{\partial x^{2}} + U(x) \psi(x,t)  $$

We have an electron in a well which is initially located around $x=5$ moving with momentum $k_{0}$,

$$ \psi(x,t=0) = e^{-\frac{1}{2} (\frac{x - 5}{\sigma_{0}})^{2}} e^{i k_{0} x} $$

Our Hamiltonian assumes the form 

$$\hat H  = -\frac{\hbar^2}{2m} \Delta + \hat U(x) $$

$$ H = -\nabla^{2} + U(x) $$



### Method 1

Define 

$$ M = \left(1 - \frac{dt}{2i} H \right)^{-1} \left(1 + \frac{dt}{2i} H \right) $$

So we have

$$ \psi(x,t+dt) = M \cdot \psi(x,t) $$

### Method 2

For a time-independent Hamiltonian the analytic solution to the TDSE is given by

$$\psi(t)=e^{-\frac{i}{\hbar} H t} \psi_{0}$$

where $\psi_{0}$ is the initial state and $\hat{H}$ is the Hamiltonian matrix. A simple way of time stepping in this case is then NumPy's expm() function, which is used as a propagator. The function of a propagator is the propagation (in time) of the function it is applied to. In this case the solution is analytic, and thus the order of accuracy is infinity.




```python
import os
import time
import warnings
import numpy as np
from scipy import sparse
import matplotlib.pyplot as plt
from matplotlib import animation, rc
from scipy.sparse import linalg as ln
from matplotlib.animation import FuncAnimation
from IPython.display import display, clear_output
from scipy.linalg import expm, sinm, cosm

warnings.filterwarnings('ignore')

plt.rc('font', family='serif')
plt.rc('xtick',labelsize=18)
plt.rc('ytick',labelsize=18)
%config InlineBackend.figure_format = 'retina'
```

```python
def gaussian(x, σ, k0, x0=5):
    return np.exp(-0.5 * ((x - x0) / σ) ** 2) * np.exp(1j * k0 * x) 


def free(npts):
    "Free particle."
    return np.zeros(npts)


def barrier(npts,v0,thickness, offset=0):
    "Barrier potential (also a well)"
    v = free(npts)
    v[int(offset+npts/2):int(offset+npts/2+thickness)] = v0
    return v


def triangle(npts, vmax, width, offset=0):
    "Triangle potential"
    v = free(npts)
    A = (2*vmax)/npts  # slope 
    up_V = (np.arange(int(offset+(npts/2-width/2)),int(offset+npts/2))  - int(offset+(npts/2-width/2)))*A
    down_V = np.flip(np.arange(int(offset+npts/2), int(offset+(npts/2+width/2))) - int(offset+(npts/2)))*A
    v[int(offset+(npts/2-width/2)):int(offset+npts/2)] = up_V
    v[int(offset+npts/2):int(offset+npts/2+width/2)] = down_V
    return v


def half_triangle(npts, vmax, width, offset=0):
    "half triangle potential"
    v = free(npts)
    A = (2*vmax)/npts  # slope 
    up_V = (np.arange(int(offset+(npts/2-width/2)),int(offset+npts/2))  - int(offset+(npts/2-width/2)))*A
    down_V = np.flip(np.arange(int(offset+npts/2), int(offset+(npts/2+width/2))) - int(offset+(npts/2)))*A
    v[int(offset+(npts/2-width/2)):int(offset+npts/2)] = up_V
    v[int(offset+npts/2):int(offset+npts/2+width/2)] = up_V + up_V.max()
    return v


def SHO(npts, k, width, offset):
    """Harmonic oscillator potential"""
    v = free(npts)
    half = np.arange(int(offset+npts/2), int(offset+(npts/2+width/2))) - int(offset+(npts/2))
    up_curve = 0.5*k*half**2
    down_curve = np.flip(up_curve)
    v[int(offset+(npts/2-width/2)):int(offset+npts/2)] = down_curve
    v[int(offset+npts/2):int(offset+npts/2+width/2)] = up_curve
    v[int(offset+(npts/2-width/2)):int(offset+npts/2+width/2)] -=  v.max()
    return v
    
```

```python
h = 1
m = 1
Δx = 1
Nx = 1400
σ0 = 30
v0 = 0.2
x0 = round(Nx/2) - 10*σ0 # Time shift
k0 = np.pi / 20
# k0 = (m*v0)/h


""" 1) Space discretization """
x = np.linspace(0,Nx,Nx)
xmin = x.min()
xmax = x.max()


""" 2) Initialization of the wave function to Gaussian wave packet """
Ψ0 = gaussian(x=x, σ=σ0, k0=k0, x0=x0)
Ψ0_R = Ψ0.real
Ψ0_I = Ψ0.imag
ΨP = Ψ0_R**2 + Ψ0_I**2
P = ΨP.sum()
nrm = np.sqrt(P)
Ψ0 /= nrm
ΨP /= P


""" 3) Setting up the potential"""
V = barrier(npts=len(x), v0=0.015, thickness=25, offset = 0)
# V = -triangle(npts=len(x), vmax=0.05, width=1400, offset=0)
# V1 = -half_triangle(npts=len(x), vmax=0.05, width=600, offset=300)
# V2 = -np.flip(half_triangle(npts=len(x), vmax=0.05, width=600, offset=300))
# V3 = SHO(npts=len(x), k=2e-7, width=1400, offset=0) 
Vabsmax = np.abs(V.min()) if (V.max() == 0) else V.max()
Δt = h/((h**2/(m*Δx**2))+(Vabsmax/2)) * 10


""" 4) Creating the Hamiltonian """
H_diag = (h**2/(2*m*Δx**2)) * 2 * np.ones(Nx) + V
H_non_diag = (h**2/(2*m*Δx**2)) * -1 * np.ones(Nx - 1) 
H = sparse.diags([H_diag, H_non_diag, H_non_diag], [0, 1, -1])


""" Creating the time evolution matrix Using Method 1"""
A = sparse.eye(Nx) - Δt/2.0j * H
B = sparse.eye(Nx) + Δt/2.0j * H
M = ln.inv(A).dot(B)
M = M.toarray()


""" Creating the time evolution matrix Using Method 2"""
H = H.toarray()
U = expm(-1j*(H/h)*Δt)


ymax = 1.5*Ψ0_R.max()
Efac = 1.5*(Ψ0_R).max()/2.0/Vabsmax

E = (h**2/(2.0*m))*(k0**2+0.5/σ0**2)
KE = (h**2*k0**2)/(2*m)
print("Wavepacket Energy: {:.4f}".format(E))
print("Potential: {:.4f}".format(Vabsmax))
print("Δt = {:.4f}".format(Δt))
print("sum(ΨP) = {:.4f}".format(ΨP.sum()))
```

```python
plt.figure(figsize=(16,8))
plt.axis([xmin,xmax,-ymax,ymax])
plt.title("Initial Electron Wave Function", fontsize=20)
plt.plot(x, 6*ΨP, 'k' , lw=2, label=r'$ | \Psi(x,0) |^{2} $')
plt.plot(x, Ψ0_R, 'r', alpha=0.7, label=r'$ \Psi(x,0) $ Real')
plt.plot(x, Ψ0_I, 'b', alpha=0.7, label=r'$ \Psi(x,0) $ Imag')
plt.plot(x, Efac*V, ':k', lw=2, alpha=0.2, label=r'V')
plt.axhline(E*Efac, color='g',label='Energy', zorder=1)
plt.fill_between(x,Efac*V, facecolor='y', alpha=0.4, zorder=0)
plt.xlabel("x", fontsize=18)
plt.legend(prop={'size': 14}, loc="lower right")
plt.show()
```

```python
class WavePacket:
    def __init__(self, Ψ, M, U):
        self.Ψ = Ψ
        self.M = M
        self.U = U
    
    def evolve_U(self):
        self.Ψ = self.U.dot(self.Ψ)
        ΨP = self.Ψ.real**2 + self.Ψ.imag**2
        return self.Ψ, ΨP
    
    def evolve_M(self):
        self.Ψ = self.M.dot(self.Ψ)
        ΨP = self.Ψ.real**2 + self.Ψ.imag**2
        return self.Ψ, ΨP
```

```python
wave_packet = WavePacket(Ψ=Ψ0, M=M, U=U)

Nt = 1000

ΨR_arr = np.zeros((Nt, Nx))
ΨI_arr = np.zeros((Nt, Nx))
ΨP_arr = np.zeros((Nt, Nx))

start = time.time() 

for i in range(0, Nt):
    Ψ, ΨP = wave_packet.evolve_U()
    ΨR_arr[i] = Ψ.real
    ΨI_arr[i] = Ψ.imag
    ΨP_arr[i] = ΨP
    if (i % 50) == 0:
        clear_output(wait=True)
        print("Norm = {}".format(ΨP.sum()))
        print("Timestep {} / {}".format(i, Nt))
    
end = time.time()
print("\n{:.4f} seconds".format(end-start))
```
